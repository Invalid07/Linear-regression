```python
for col_name in col:
    if df[col_name].dtypes =='object':
        #object
        df[col_name] = df[col_name].fillna(df[col_name].mode()[0])
    else:
        #numeric
        df[col_name] = df[col_name].fillna(df[col_name].mean())
```
